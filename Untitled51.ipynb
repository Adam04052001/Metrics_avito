**Создадим когорты**

```python
orders['order_date'] = pd.to_datetime(orders['order_date']) 
```

```python
orders['order_month'] = orders['order_date'].dt.strftime('%Y-%m') 
```

```python
cohorts= orders.groupby('customer_id',as_index=False)['order_month'].min()
cohorts.rename(columns = {'order_month':'cohort'},inplace=True)
display(cohorts)
```

```python
data = pd.merge(orders,cohorts , how='left', on='customer_id')
```

```python
data['order_date'] = pd.to_datetime(data['order_date'],infer_datetime_format=True)
data['order_month'] = data['order_date'].dt.strftime('%Y-%m') 
```

```python
data
```

```python
def cohort_period(data_full):
    """
    Creates column CohortPeriod
    """
    data_full['CohortPeriod'] = np.arange(len(data_full))
    return data_full
```

```python
data.head(0)
```

```python
cohorts_group = data.groupby(['cohort', 'order_month']).agg({'customer_id': pd.Series.nunique})
cohorts_group.rename(columns={'customer_id': 'TotalUsers',
                        'id': 'TotalOrders'}, inplace=True)
cohorts_group = cohorts_group.groupby(level=0).apply(cohort_period)
cohorts_group.reset_index(inplace=True)
cohorts_group.set_index(['cohort', 'CohortPeriod'], inplace=True) 
cohort_group_size = cohorts_group['TotalUsers'].groupby(level=0).first()
user_retention = cohorts_group['TotalUsers'].unstack(0).divide(cohort_group_size, axis=1)
```

```python
cohorts_group = data.groupby(['cohort', 'order_month']).agg({'customer_id': pd.Series.nunique})
cohorts_group.rename(columns={'customer_id': 'TotalUsers', 'id_x': 'TotalOrders'}, inplace=True)
cohorts_group = cohorts_group.groupby(level=0).apply(cohort_period)
cohorts_group.reset_index(inplace=True)
cohorts_group.set_index(['cohort', 'CohortPeriod'], inplace=True) 
cohort_group_size = cohorts_group['TotalUsers'].groupby(level=0).first()
user_retention = cohorts_group['TotalUsers'].unstack(0).divide(cohort_group_size, axis=1)
```

```python
user_retention.T
```

```python
sns.set(style='white')
plt.figure(figsize=(25, 16))
plt.title('Cohorts: User Retention')
sns.heatmap(user_retention.T, mask=user_retention.T.isnull(), annot=True, fmt='.0%');
```

**После того, как была получена когорта, можно переходить к их анализу**

Нам надо проанализировать данные за 2016 и 2017 года, поэтому обрежем данные data по нужному значению атрибута даты, построим когорту и проанализруем её 

```python
orders_2016 = orders[orders['order_date'] >= '2016-01-01']
```

```python
cohorts = orders_2016.groupby('customer_id', as_index=True).min()['order_date']
cohorts = pd.DataFrame(cohorts)
cohorts.reset_index(level=0, inplace=True)
cohorts['order_date'] = pd.to_datetime(cohorts['order_date'],infer_datetime_format=True)
cohorts['Cohort'] = cohorts['order_date'].dt.strftime('%Y-%m')
cohorts = cohorts.drop(['order_date'], axis = 1)
```

```python
data_2016 = pd.merge(orders_2016, cohorts, how='left', on='customer_id')
```

```python
data_2016['order_date'] = pd.to_datetime(data_2016['order_date'],infer_datetime_format=True)
data_2016['order_month'] = data_2016['order_date'].dt.strftime('%Y-%m') 
```

```python
cohorts_group_2016 = data_2016.groupby(['Cohort', 'order_month']).agg({'customer_id': pd.Series.nunique})
cohorts_group_2016.rename(columns={'customer_id': 'TotalUsers',
                        'id': 'TotalOrders'}, inplace=True)
cohorts_group_2016 = cohorts_group_2016.groupby(level=0).apply(cohort_period)
cohorts_group_2016.reset_index(inplace=True)
cohorts_group_2016.set_index(['Cohort', 'CohortPeriod'], inplace=True) 
cohort_group_2016_size = cohorts_group_2016['TotalUsers'].groupby(level=0).first()
user_retention_2016 = cohorts_group_2016['TotalUsers'].unstack(0).divide(cohort_group_2016_size, axis=1)
```

```python
user_retention_2016.T
```

```python
sns.set(style='white')
plt.figure(figsize=(16, 10))
plt.title('Cohorts: User Retention')
sns.heatmap(user_retention_2016.T, mask=user_retention_2016.T.isnull(), annot=True, fmt='.0%');
```

Анализируя когорты 2016 года и их изменение в 2017 году, можно заметить диаогонали стали куда светлее и принимают более высокое процентное значение. То есть повысилась лояльность клиентов, и они стали чаще приобретать продукты компании, что может свидетельствовать, что сейчас лучше, чем было 

Также по когортам квадртата за 2017 год, что retention тоже куда выше, чем у квадрата за 2016 год, что также свиделеьтсвует, что в 2017 году стало лучше, чем в 2016. Однако заметим, что диагональ за 2017 года не полная, отсутвуют значения, что не даёт возможности спокойно заявить об однозначном улучшении ситуации в 2017 году.
